##Mendeteksi Outlier

```python
boxplot = (data['absent_hours'], data['age'], data['length_service'])
labels = ('absent_hours','age','length_service')
plt.boxplot(boxplot, labels=labels)
plt.show()
```

```python
#Removing Outliers on Age
sorts=data['age'].sort_values()
q1=sorts.quantile(0.25)
q3=sorts.quantile(0.75)
iqr = q3 - q1
upper=q1-1.5*iqr
lower=q3+1.5*iqr
data['age']=sorts[~((sorts<(upper))  |  (sorts>(lower)))]
sns.boxplot(data['age'])
```

```python
#Removing Outliers on Length Service
sorts=data['length_service'].sort_values()
q1=sorts.quantile(0.25)
q3=sorts.quantile(0.75)
iqr = q3 - q1
upper=q1-1.5*iqr
lower=q3+1.5*iqr
data['length_service']=sorts[~((sorts<(upper))  |  (sorts>(lower)))]
sns.boxplot(data['length_service'])
```

```python
#Removing Outliers on Absent Hours
sorts=data['absent_hours'].sort_values()
q1=sorts.quantile(0.25)
q3=sorts.quantile(0.75)
iqr = q3 - q1
upper=q1-1.5*iqr
lower=q3+1.5*iqr
data['absent_hours']=sorts[~((sorts<(upper))  |  (sorts>(lower)))]
sns.boxplot(data['absent_hours'])
```

#Modeling

```python
data.describe()
```

```python
##Mendefinisikan Variabel
x = data[['age']]
y = data[['absent_hours']]
```

```python
##Split Data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
 # create linear regression object
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
##Evaluate
y_pred = reg.predict(x_test)
```

```python
#r2_score
from sklearn.metrics import r2_score
# MAE
from sklearn.metrics import mean_absolute_error
# MSE
from sklearn.metrics import mean_squared_error
# RMSE
from sklearn.metrics import mean_squared_error
```

```python
print ("r2:",r2_score(y_test,y_pred))
print ("MAE : ",mean_absolute_error(y_test,y_pred))
print ("MSE: ",mean_squared_error(y_test,y_pred))
print ("RMSE: ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

**Interpretasi**

MAE (rata-rata error):
Bahwa prediksi Absent Hours akan error + 20.332224425264855 atau - 20.332224425264855 dari yang seharusnya. 

RMSE:
Bahwa prediksi Absent Hours akan error + 25.344387380381598 atau - 25.344387380381598 dari waktu aslinya.

r2:
Diketahui bahwa akurasi model regresi yang terbentuk ialah sebesar 71%. Hal itu menandakan bahwa model regresi yang tebentuk sudah cukup kuat. Selain itu, bisa dijelaskan bahwa variabel independen mempengaruhi variabel dependen sebesar 71% dan 29% lainnya dipengaruhi variabel lain di luar model.
